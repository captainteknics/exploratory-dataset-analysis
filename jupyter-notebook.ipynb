```python
# Step 1: Load the Dataset
import pandas as pd
```

```python
# Load the dataset into a DataFrame
file_path = 'socialmedia-disaster-tweets-DFE.csv'

 # Specify the encoding explicitly
df = pd.read_csv(file_path, encoding='latin1')
```

```python
# Display the first few rows of the DataFrame to understand its structure
print(df.head())
```

### Step 2: Exploratory Data Analysis (EDA)

```python
# General statistics
print("General Statistics:")
print(df.describe())
```

```python
# Check the column names to identify the correct target variable
print("\nColumn Names:")
print(df.columns)
```

## b) visualization of main variables.

```python
# Visualization of main variable 'choose_one'
import matplotlib.pyplot as plt

# Visualizing the distribution of 'choose_one' (relevance to disaster)
plt.figure(figsize=(8, 6))
df['choose_one'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.xlabel('Relevance to Disaster (0: Non-Relevant, 1: Relevant)')
plt.ylabel('Frequency')
plt.title('Distribution of Relevance to Disaster')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
```

```python
# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Text Data')
plt.show()
```
